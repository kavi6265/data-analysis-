```python
import pandas as pd

# Path to your dataset on the desktop
# For Windows


# For macOS/Linux
# file_path = '/Users/YourUsername/Desktop/my_dataset.csv'

# Load the dataset into a pandas DataFrame
df = pd.read_csv("D:/gas.Excel.csv")

# Display the first few rows of the dataset
df
```

```python
df["PM2.5"].fillna(df["PM2.5"].mean(), inplace = True)
df["PM10"].fillna(df["PM10"].mean(), inplace = True)
df["NO"].fillna(df["NO"].mean(), inplace = True)
df["NO2"].fillna(df["NO2"].mean(), inplace = True)
df["NOx"].fillna(df["NOx"].mean(), inplace = True)
df["NH3"].fillna(df["NH3"].mean(), inplace = True)
df["CO"].fillna(df["CO"].mean(), inplace = True)
df["SO2"].fillna(df["SO2"].mean(), inplace = True)
df["O3"].fillna(df["O3"].mean(), inplace = True)
df["Benzene"].fillna(df["Benzene"].mean(), inplace = True)
df["Toluene"].fillna(df["Toluene"].mean(), inplace = True)
df["Xylene"].fillna(df["Xylene"].mean(), inplace = True)
df["AQI"].fillna(df["AQI"].mean(), inplace = True)
```

```python
df
```

```python
df["AQI_Bucket"].fillna("Good",inplace = True)
df
```

```python
df
```

```python
# Save the DataFrame 'df' to an Excel file
df.to_excel("D:/analysis_dataset.xlsx.xlsx", index=False, engine='openpyxl')

```

```python
print(df.isnull().sum())
```

```python
df_melted = pd.melt(df, id_vars=["City"], value_vars=['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene', 'AQI'], 
                    var_name="Element", value_name="Value")

# Create a figure
plt.figure(figsize=(25, 8))

# Barplot
sns.barplot(x='City', y='Value', hue='Element', data=df_melted)


# Adding title and labels
plt.title("City Vs Element Concentration")
plt.xlabel("City")
plt.ylabel("")

# Rotate x-axis labels for readability
plt.xticks(rotation=45)

# Set a fixed location for the legend
  # You can change the location to suit your plot

# Show the plot
plt.tight_layout()
plt.show()
```
